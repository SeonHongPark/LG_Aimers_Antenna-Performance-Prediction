#평가지표

```python
# 평가지표
from sklearn.metrics import mean_squared_error
def lg_nrmse(gt, preds):
    all_nrmse = []
    temp = len(gt.columns)
    gt = np.array(gt)
    preds = np.array(preds)
    for idx in range(temp): 
        rmse = mean_squared_error(gt[:,idx], preds[:,idx], squared=False)
        nrmse = rmse/np.mean(np.abs(gt[:,idx]))
        all_nrmse.append(nrmse)
    score = 1.2 * np.sum(all_nrmse[:8]) + 1.0 * np.sum(all_nrmse[8:14])
    print(score)
    return score
```

# k-fold - lgbm

```python
from sklearn.model_selection import KFold
import lightgbm 
kf = KFold(n_splits=10 , random_state = 42 , shuffle = True)
i = 1
lgbm_model_save = []
nrmse_score = []

for train_index, test_index in kf.split(train_x):
  
  print(train_index)
  X_train, X_test = train_x.iloc[train_index], train_x.iloc[test_index]
  Y_train, Y_test = train_y.iloc[train_index], train_y.iloc[test_index]

  lgbm_model= MultiOutputRegressor(lightgbm.LGBMRegressor(random_seed = 42,
                                                          num_leaves = 681, colsample_bytree = 0.7732774027764344, 
                                                          reg_alpha = 0.3718384437002339, reg_lambda = 5.467302098418594, 
                                                          max_depth = 15, learning_rate = 0.0037878535825736077, n_estimators= 2994, 
                                                          min_child_samples= 29, subsample= 0.5855365968947291)).fit(X_train, Y_train)

  lgbm_model_save.append(lgbm_model)
  temp = lg_nrmse(Y_test, pd.DataFrame(lgbm_model.predict(X_test)))
  nrmse_score.append(temp)
  print(f"{i}번째 폴드 끝")
  i += 1
print(sum(nrmse_score)/10)
```

```python
lgbm_result = pd.DataFrame(lgbm_model_save[0].predict(test_x)) +pd.DataFrame(lgbm_model_save[1].predict(test_x)) +\
pd.DataFrame(lgbm_model_save[2].predict(test_x)) +pd.DataFrame(lgbm_model_save[3].predict(test_x)) +\
pd.DataFrame(lgbm_model_save[4].predict(test_x)) +pd.DataFrame(lgbm_model_save[5].predict(test_x)) +\
pd.DataFrame(lgbm_model_save[6].predict(test_x)) +pd.DataFrame(lgbm_model_save[7].predict(test_x)) +\
pd.DataFrame(lgbm_model_save[8].predict(test_x)) +pd.DataFrame(lgbm_model_save[9].predict(test_x))
```

```python
lgbm_result = lgbm_result / 10
```

```python
from sklearn.model_selection import KFold
from sklearn.ensemble import ExtraTreesRegressor
kf = KFold(n_splits=3 , random_state = 42 , shuffle = True)
i = 1
et_model_save = []
nrmse_score = []

for train_index, test_index in kf.split(train_x):
  
  print(train_index)
  X_train, X_test = train_x.iloc[train_index], train_x.iloc[test_index]
  Y_train, Y_test = train_y.iloc[train_index], train_y.iloc[test_index]

  et_model= MultiOutputRegressor(ExtraTreesRegressor(random_state=42,n_estimators = 310, max_depth = 18,
                                                     min_samples_split = 17, min_samples_leaf = 2)).fit(X_train, Y_train)

  et_model_save.append(et_model)
  temp = lg_nrmse(Y_test, pd.DataFrame(et_model.predict(X_test)))
  nrmse_score.append(temp)
  print(f"{i}번째 폴드 끝")
  i += 1
print(sum(nrmse_score)/10)
```
